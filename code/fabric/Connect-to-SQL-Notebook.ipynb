```python

from notebookutils import mssparkutils
from pyspark.sql import SparkSession
# JDBC Connection
serverName = "<YOUR_SERVER_NAME>.database.windows.net>"
database = "<YOUR_DATABASE_NAME>"
dbPort = 1433
jdbcURL = f"jdbc:sqlserver://{serverName}:{dbPort};database={database};encrypt=true;trustServerCertificate=false;hostNameInCertificate=*.database.windows.net;loginTimeout=30;"
# Get Entra token
token = mssparkutils.credentials.getToken("https://database.windows.net/")
connectionProperties = {
    "accessToken": token,
    "driver": "com.microsoft.sqlserver.jdbc.SQLServerDriver"
}
# Read SQL Server table
df = spark.read.jdbc(
    url=jdbcURL,
    table="dbo.SalesPrivate",
    properties=connectionProperties
)
display(df)
```

```python
df.write.mode("overwrite").saveAsTable("<YOUR_DATABASE_TABLE_NAME>")
```
